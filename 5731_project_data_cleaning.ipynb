### Pre Processing the Data 

* removing punctuations and special characters
* lower all the text
* remove stop words
* remove numbers
* stemming
* lemmatizing

```python
# Write your code here
import nltk, re, string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from textblob import Word
nltk.download('stopwords')
nltk.download("wordnet")
nltk.download('omw-1.4')
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer

def remove_non_english_words(text):
    """
    Removes non-English words from the given text and returns the result.
    """
    pattern = r'[^\x00-\x7F]+' # matches any non-ASCII characters
    english_text = re.sub(pattern, ' ', text) # replaces non-ASCII characters with spaces
    words = english_text.split()
    english_words = [word for word in words if re.match(r'^[a-zA-Z]+$', word)] # matches only English words
    return ' '.join(english_words)

def tokenize_sentence(s):
  st = PorterStemmer()
  lemmatizer = WordNetLemmatizer()
  tokenizer = RegexpTokenizer(r'\w+')
  tokens = tokenizer.tokenize(s) 
  stem_words=[st.stem(w) for w in tokens]
  lemma_words=[lemmatizer.lemmatize(w) for w in stem_words]
  return " ".join(tokens) 

def remove_puntutation_and_special_chars(s):
  """
    params1: string containing sentence
    return: string with no punctuations
  """
  s = s.replace('{html}',"") 
  s = re.sub(r'http\S+', '',s)
  s = re.sub(r'@\w+', '', s)
  cleanr = re.compile('<.*?>')
  cleantext = re.sub(cleanr, '', s)
  # return cleantext
  return "".join([x for x in cleantext if x not in string.punctuation])

def remove_nums(s):
  # for x in range(0, 10):
  #   s = s.replace(str(x), '')
  # return s
  return re.sub(r'[0-9]+', '', str(s))

def remove_stopwords(s):
  stop = stopwords.words("english")
  return " ".join(x for x in s.split() if len(x)>2 if x not in stop)

def lower_all(s):
  return " ".join(x.lower() for x in s.split())


df["Full Text"] = df["Full Text"].astype(str)
df["text_modified"] = df["Full Text"].replace("\n","")
operations = [remove_puntutation_and_special_chars, lower_all, remove_stopwords]

for x in range(len(df)):
  df.iloc[x, -1] = remove_puntutation_and_special_chars(df.iloc[x, -1])
  df.iloc[x, -1] = lower_all(df.iloc[x, -1])
  df.iloc[x, -1] = remove_stopwords(df.iloc[x, -1])
  df.iloc[x, -1] = remove_nums(df.iloc[x, -1])
  df.iloc[x, -1] = tokenize_sentence(df.iloc[x, -1])
  df.iloc[x, -1] = remove_non_english_words(df.iloc[x, -1])
# df["text_modified"].apply(remove_puntutation_and_special_chars)
train_df = df.copy()
train_df.head()
```
